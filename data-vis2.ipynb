```python
# categories
cat = {
    0 : "No_financial_services",
    1 : "Other_only",
    2 : "Mm_only",
    3 : "Mm_plus"
}
# group and get relevant data
services_age = tan.groupby(["mobile_money_classification", "english_literate"])["ID"].count().reset_index()
font = {
    'weight' : 'normal',
    'size'   : 15
       }
# some settings to make sure client can see text.
plt.rc('font', **font)
plt.figure(figsize=(20,10))
for x in services_age["mobile_money_classification"].unique():
    plt.plot( 'english_literate', 'ID', data=services_age[services_age.english_literate==x],
             marker='', linewidth=3, label="Label for {}".format(cat[x])
            )
plt.title("Relationship between financial services accessed (non-mobile, mobile, both) and english_literate")
plt.legend()
plt.show()
```
